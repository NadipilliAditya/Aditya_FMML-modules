### Slope & Derivatives

Creating a function for the derivative of $f(x)$ called df(x)

````python
def df(x):
    return 2 * x + 1
````

### Gradient Descent

````python
# Gradient Descent
new_x = 3
previous_x = 0
step_multiplier = 0.1
precision = 0.00001

x_list = [new_x]
slope_list = [df(new_x)]

for n in range(500):
    previous_x = new_x
    gradient = df(new_x)  # [REQUIRED]: Gradient is the derivative of f(x) at new_x
    new_x = previous_x - step_multiplier * gradient  # [REQUIRED]: Update rule for x

    step_size = abs(new_x - previous_x)
    # print(step_size)

    x_list.append(new_x)
    slope_list.append(df(new_x))  # [REQUIRED]: Store the new slope at new_x

    if step_size < precision:
        print('Loop ran this many times:', n)
        break

print('Local minimum occurs at:', new_x)
print('Slope or df(x) value at this point is:', df(new_x))
print('f(x) value or cost at this point is:', f(new_x))
````

````python
# Superimpose the gradient descent calculations on plot

plt.figure(figsize=[20, 5])

# 1 Chart: Cost function
plt.subplot(1, 3, 1)

plt.xlim(-3, 3)
plt.ylim(0, 8)

plt.title('Cost function', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('f(x)', fontsize=16)

plt.plot(x_1, f(x_1), color='blue', linewidth=3, alpha=0.8)

values = np.array(x_list)
plt.scatter(x_list, f(values), color='red', s=100, alpha=0.6)

# 2 Chart: Derivative
plt.subplot(1, 3, 2)

plt.title('Slope of the cost function', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('df(x)', fontsize=16)
plt.grid()
plt.xlim(-2, 3)
plt.ylim(-3, 6)

plt.plot(x_1, df(x_1), color='skyblue', linewidth=5, alpha=0.6)
plt.scatter(x_list, slope_list, color='red', s=100, alpha=0.5)

# 3 Chart: Derivative (Close Up)
plt.subplot(1, 3, 3)

plt.title('Gradient Descent (close up)', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.grid()
plt.xlim(-0.55, -0.2)
plt.ylim(-0.3, 0.8)

plt.plot(x_1, df(x_1), color='skyblue', linewidth=6, alpha=0.8)
plt.scatter(x_list, slope_list, color='red', s=300, alpha=0.6)


plt.show()
````

## Gradient Descent

````python
def gradient_descent(derivative_func, initial_guess, multiplier=0.02, precision=0.001,
                    max_iter=300):
    new_x = initial_guess
    x_list = [new_x]
    slope_list = [derivative_func(new_x)]

    for n in range(max_iter):
        previous_x = new_x
        gradient = derivative_func(previous_x)
        new_x = previous_x - multiplier * gradient

        step_size = abs(new_x - previous_x)
        x_list.append(new_x)
        slope_list.append(derivative_func(new_x))

        if step_size < precision:
            break
    return new_x, x_list, slope_list
````

````python
import numpy as np
import matplotlib.pyplot as plt

# Define the function and its derivative
def g(x):
    return x**2 * np.sin(x)  # Example function g(x)

def dg(x):
    return 2*x*np.sin(x) + x**2*np.cos(x)  # Derivative of g(x)

# Gradient Descent Implementation
def gradient_descent(derivative_func, initial_guess, multiplier, precision, max_iter):
    x = initial_guess
    steps = [x]  # To track progress
    for i in range(max_iter):
        prev_x = x
        gradient = derivative_func(prev_x)
        x = prev_x - multiplier * gradient  # Gradient descent step
        steps.append(x)
        if abs(x - prev_x) < precision:
            break
    return np.array(steps)

# Parameters
n = 100
low_gamma = gradient_descent(derivative_func=dg, initial_guess=3,
                             multiplier=0.0005, precision=0.0001, max_iter=n)

mid_gamma = gradient_descent(derivative_func=dg, initial_guess=3,
                              multiplier=0.001, precision=0.0001, max_iter=n)  # [REQUIRED]

high_gamma = gradient_descent(derivative_func=dg, initial_guess=3,
                               multiplier=0.002, precision=0.0001, max_iter=n)  # [REQUIRED]

# Experiment with insane_gamma
insane_gamma = gradient_descent(derivative_func=dg, initial_guess=1.9,
                                 multiplier=0.25, precision=0.0001, max_iter=n)

# Visualization
plt.figure(figsize=(12, 8))

# Plot low_gamma
plt.plot(low_gamma, g(low_gamma), label="Low Learning Rate (0.0005)", marker="o")

# Plot mid_gamma
plt.plot(mid_gamma, g(mid_gamma), label="Mid Learning Rate (0.001)", marker="x")  # [REQUIRED]

# Plot high_gamma
plt.plot(high_gamma, g(high_gamma), label="High Learning Rate (0.002)", marker="s")  # [REQUIRED]

# Plot insane_gamma
plt.plot(insane_gamma, g(insane_gamma), label="Insane Learning Rate (0.25)", marker="D")

plt.title("Gradient Descent with Different Learning Rates")
plt.xlabel("Steps")
plt.ylabel("g(x) Value")
plt.legend()
plt.grid()
plt.show()
````

````python
# Gradient Descent Function Modified to Track Costs
def gradient_descent_with_cost(derivative_func, func, initial_guess, multiplier, precision, max_iter):
    x = initial_guess
    steps = [x]  # To track progress
    costs = [func(x)]  # To track the cost (value of g(x))
    for i in range(max_iter):
        prev_x = x
        gradient = derivative_func(prev_x)
        x = prev_x - multiplier * gradient  # Gradient descent step
        steps.append(x)
        costs.append(func(x))
        if abs(x - prev_x) < precision:
            break
    return np.array(steps), np.array(costs)

# Run Gradient Descent for Different Learning Rates
n = 100
low_steps, low_costs = gradient_descent_with_cost(dg, g, initial_guess=3, multiplier=0.0005, precision=0.0001, max_iter=n)
mid_steps, mid_costs = gradient_descent_with_cost(dg, g, initial_guess=3, multiplier=0.001, precision=0.0001, max_iter=n)
high_steps, high_costs = gradient_descent_with_cost(dg, g, initial_guess=3, multiplier=0.002, precision=0.0001, max_iter=n)
insane_steps, insane_costs = gradient_descent_with_cost(dg, g, initial_guess=1.9, multiplier=0.25, precision=0.0001, max_iter=n)

# Plot Reduction in Cost for Each Iteration
plt.figure(figsize=(12, 8))

# Plot for Low Learning Rate
plt.plot(range(len(low_costs)), low_costs, label="Low Learning Rate (0.0005)", marker="o")

# Plot for Mid Learning Rate
plt.plot(range(len(mid_costs)), mid_costs, label="Mid Learning Rate (0.001)", marker="x")

# Plot for High Learning Rate
plt.plot(range(len(high_costs)), high_costs, label="High Learning Rate (0.002)", marker="s")

# Plot for Insane Learning Rate
plt.plot(range(len(insane_costs)), insane_costs, label="Insane Learning Rate (0.25)", marker="D")

# Add Labels, Title, and Legend
plt.title("Cost Reduction Over Iterations for Different Learning Rates")
plt.xlabel("Iterations")
plt.ylabel("Cost (g(x))")
plt.legend()
plt.grid()
plt.show()
````

### Example 4 - Data Viz with 3D Charts

#### Minimise $$f(x, y) = \frac{1}{3^{-x^2 - y^2} + 1}$$

Minimise $$f(x, y) = \frac{1}{r + 1}$$ where $r$ is $3^{-x^2 - y^2}$

````python
import numpy as np
import matplotlib.pyplot as plt

# Example Function and Derivative
def g1(x):
    return x**2  # First example: g1(x) = x^2

def dg1(x):
    return 2 * x  # Derivative: dg1(x)/dx = 2x

def g2(x):
    return x**3 - 3 * x  # Second example: g2(x) = x^3 - 3x

def dg2(x):
    return 3 * x**2 - 3  # Derivative: dg2(x)/dx = 3x^2 - 3

# Gradient Descent Function Modified to Track Costs
def gradient_descent_with_cost(derivative_func, func, initial_guess, multiplier, precision, max_iter):
    x = initial_guess
    steps = [x]
    costs = [func(x)]
    for i in range(max_iter):
        prev_x = x
        gradient = derivative_func(prev_x)
        x = prev_x - multiplier * gradient
        steps.append(x)
        costs.append(func(x))
        if abs(x - prev_x) < precision:
            break
    return np.array(steps), np.array(costs)

# Parameters
n = 100  # Max iterations
learning_rate = 0.01  # Common learning rate for comparison

# Run Gradient Descent for Both Examples
steps1, costs1 = gradient_descent_with_cost(dg1, g1, initial_guess=3, multiplier=learning_rate, precision=0.0001, max_iter=n)
steps2, costs2 = gradient_descent_with_cost(dg2, g2, initial_guess=3, multiplier=learning_rate, precision=0.0001, max_iter=n)

# Plot Reduction in Cost for Both Examples
plt.figure(figsize=(12, 6))

# Plot for First Example
plt.plot(range(len(costs1)), costs1, label="g(x) = x^2", marker="o")

# Plot for Second Example
plt.plot(range(len(costs2)), costs2, label="g(x) = x^3 - 3x", marker="x")

# Add Labels, Title, and Legend
plt.title("Cost Reduction Over Iterations")
plt.xlabel("Iterations")
plt.ylabel("Cost (g(x))")
plt.legend()
plt.grid()
plt.show()
````

### Example 5 - Working with data & a real cost function

#### Mean Squared Error: a cost function for regression problems

#### $$RSS = \sum_{i=1}^{n} \big( y^{(i)} - h_\theta x^{(i)} \big)^2 $$

````python
def grad(x, y, thetas):
    m = len(y)  # Number of data points
    predictions = thetas[0] + thetas[1] * x  # Linear model: y_hat = theta_0 + theta_1 * x
    error = predictions - y
    grad_0 = (1/m) * np.sum(error)  # Gradient for theta_0
    grad_1 = (1/m) * np.sum(error * x)  # Gradient for theta_1
    return np.array([grad_0, grad_1])
````
